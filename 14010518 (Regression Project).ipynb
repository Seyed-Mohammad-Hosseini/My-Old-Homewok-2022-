```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import matplotlib_inline
```

```python
df=pd.read_csv('houseprice.csv')
df.head(10)
```

```python
#اینجا مساحت خانه را از نوع ابجکت به فلوت تبدیل میکنیم
df['Area'] = pd.to_numeric(df['Area'], errors='coerce')
```

```python
df.dtypes
```

```python
df.shape
df.describe()
```

```python
#داده های پرت رو حذف میکنیم
df.mask(df.sub(df.mean()).div(df.std()).abs().gt(3))
```

```python
#چک میکنیم که داده های پرت جایگزین شده باشه با NaN
print(df[800:820])
```

```python
#به ادرس ها لیبل می دهیم چون در رگرسیون به عدد نیاز داریم
from sklearn.preprocessing import LabelEncoder
lblAddress = LabelEncoder()
df.Address = lblAddress.fit_transform(df.Address)
```

```python
df.head()
```

```python
#ستونهایی که دوست داریم باهاشون کار کنیم رو جدا میکنیم
cdf=df[['Area', 'Room', 'Parking', 'Warehouse', 'Elevator','Address', 'Price']]
cdf.head(10)
```

```python
plt.scatter(cdf.Address,cdf.Price)
plt.show()
```

```python
cdf = cdf.dropna()
```

```python
msk= np.random.rand(len(cdf)) <0.8

test= cdf[msk]
train=cdf[~msk]
```

```python
print(cdf[800:900])
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()

train_x = np.asanyarray(train[['Area', 'Room', 'Parking', 'Warehouse','Elevator', 'Address']])
train_y = np.asanyarray(train[['Price']])

regr.fit (train_x, train_y)

print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
y_hat= regr.predict(test[['Area', 'Room', 'Parking', 'Warehouse','Elevator', 'Address']])

x = np.asanyarray(test[['Area', 'Room', 'Parking', 'Warehouse','Elevator', 'Address']])
y = np.asanyarray(test[['Price']])

print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
