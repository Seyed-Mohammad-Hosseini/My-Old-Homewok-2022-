```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart.csv')
df.head()
```

```python
df.dtypes
```

```python
df = df[pd.to_numeric(df['oldpeak'], errors='coerce').notnull()]
df['oldpeak'] = df['oldpeak'].astype('int')
```

```python
df.columns
```

```python
# داده های مورد نیاز ایکس رو جدا میکنیم تا روشون کار انجام بدیم
feature_df = df[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh',
       'exng', 'oldpeak', 'slp', 'caa', 'thall']]
X = np.asarray(feature_df)
X[0:5]
```

```python
# ایگرگ (کلاس) رو هم جدا میکنیم
df['output'] = df['output'].astype('int')
y = np.asarray(df['output'])
y [0:5]
```

# Train/Test dataset

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

# Modeling (SVM with Scikit-learn)

```python
 from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(X_train, y_train)
```

```python
# حالا که ترین رو انجام دادیم، روی داده های تستمون چکش میکنیم
yhat = clf.predict(X_test)
yhat [0:5]
```

# Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
import itertools
```

```python
# اگر  فقد اف وان اسکور رو بخوایم

from sklearn.metrics import f1_score
f1_score(y_test, yhat, average='weighted') 
```

```python
# و اگر بخوایم اندیس ژاکار رو تست کنیم

from sklearn.metrics import jaccard_score
jaccard_score(y_test, yhat,pos_label=1)
```
