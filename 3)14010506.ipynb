```python
#ضریب پیرسون رو به کمک پانداز هم میتونیم محاسبه کنیم
import pandas as pd
smartphones=pd.read_csv('C://Users//Mohammad//smartphones.csv')
num_var= smartphones.drop(['Ram', 'Capacity', 'OS', 'Company'],axis=1)
num_var
```

```python
#بدست اوردن ضریب پیرسون با پانداز
num_var=smartphones.drop([ 'OS', 'Capacity','Ram','Company'], axis=1)
num_var
```

```python
corr=num_var.corr()
corr
```

```python
#تعیین پیرسون با seaborn
import seaborn as sb
a=sb.heatmap(corr,xticklabels=corr.columns, yticklabels=corr.columns, vmin=-1, vmax=+1)
a
```

```python
#ارتباط بین متغیرهای غیرپارامتری (اسپیرمن)
cat_var=smartphones.drop(['Company','OS', 'Weight', 'inch'], axis=1)
cat_var
```

```python
from scipy.stats import spearmanr
spearman_coeeficient, _=spearmanr (cat_var.Capacity, cat_var.Ram)
spearman_coeeficient
```

```python
from scipy.stats import chi2_contingency
table=pd.crosstab (smartphones.Capacity, smartphones.Ram)
table
```

```python
chi2, p_value, dof, expected= chi2_contingency(table.values)
chi2
p_value
dof
expected
```
